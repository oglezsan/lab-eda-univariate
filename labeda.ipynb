```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/Users/oscargonzalezsanchez/Documents/ironhack/amz.csv")
```

```python
df.head()
```

```python
#Part 1: Understanding Product Categories
cat_freq = df['category'].value_counts().reset_index()
cat_freq.columns = ['Category', 'Frequency']
print(cat_freq)
```

```python
top5 = cat_freq.head(5)
print(top5)
```

```python
# Bar chart for all categories
plt.figure(figsize=(10, 6))
plt.bar(top5['Category'], top5['Frequency'], color='skyblue')
plt.xlabel('Product Category')
plt.ylabel('Frequency')
plt.title('Distribution of Products Across Categories')
plt.xticks(rotation=45)
plt.show()

# Pie chart for top 5 categories
plt.figure(figsize=(8, 8))
plt.pie(top5['Frequency'], labels=top5['Category'], autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Top 5 Product Categories')
plt.show()
```

```python
# Sports & outdoors dominates the listings
```

```python
# Part 2: Delving into Product Pricing
# Calculate mean, median, and mode for product prices
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]

print(f"Mean Price: {mean_price:.2f}")
print(f"Median Price: {median_price:.2f}")
print(f"Mode Price: {mode_price:.2f}")
```

```python
# Calculate variance, standard deviation, range, and interquartile range for product prices
variance_price = df['price'].var()
std_dev_price = df['price'].std()
range_price = df['price'].max() - df['price'].min()
iqr_price = df['price'].quantile(0.75) - df['price'].quantile(0.25)

print(f"Variance: {variance_price:.2f}")
print(f"Standard Deviation: {std_dev_price:.2f}")
print(f"Range: {range_price:.2f}")
print(f"Interquartile Range: {iqr_price:.2f}")
```

```python
import matplotlib.pyplot as plt

# Plot a histogram to visualize the distribution of product prices
plt.hist(df['price'], bins=30, color='skyblue')
plt.xlabel('price ()')
plt.ylabel('frequency')
plt.title('Distribution of Product Prices on Amazon UK')
plt.show()
```

```python
# Plot a box plot to visualize the spread and potential outliers in product pricing
plt.boxplot(df['price'], vert=False)
plt.xlabel('price ()')
plt.title('Box Plot of Product Prices on Amazon UK')
plt.show()
```

```python
#Part 3: Analyzing Product Ratings
# Calculate mean, median, and mode for product ratings
mean_rating = df['stars'].mean()
median_rating = df['stars'].median()
mode_rating = df['stars'].mode()[0]

print(f"Mean Rating: {mean_rating:.2f}")
print(f"Median Rating: {median_rating:.2f}")
print(f"Mode Rating: {mode_rating:.2f}")
```

```python
# Calculate measures of dispersion
variance_rating = df['stars'].var()
std_dev_rating = df['stars'].std()
q1 = df['stars'].quantile(0.25)
q3 = df['stars'].quantile(0.75)
iqr = q3 - q1

print(f"Variance:, {variance_rating:.2f}")
print(f"Standard Deviation:, {std_dev_rating:.2f}")
print(f"Interquartile Range: {iqr:.2f}")
```

```python
skewness_rating = df['stars'].skew()
kurtosis_rating = df['stars'].kurt()

print(f"\nSkewness: {skewness_rating:.2f}")
print(f"Kurtosis: {kurtosis_rating:.2f}")
```

```python
# Plot histogram
plt.figure(figsize=(8, 6))
sns.histplot(df['stars'], bins=10, kde=True)
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```
